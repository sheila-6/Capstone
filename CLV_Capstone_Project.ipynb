```python
import pandas as pd
import io
import requests

# Correct raw link here
url = "https://raw.githubusercontent.com/sheila-6/Capstone/main/Online%20Retail.xlsx"

response = requests.get(url)
if response.status_code == 200:
    print("✅ File downloaded successfully")
    df = pd.read_excel(io.BytesIO(response.content), engine='openpyxl')
    print(df.head())
else:
    print(f"❌ Failed to download file. Status code: {response.status_code}")

```

```python
# Step 3: Clean and rename columns
df.columns = df.columns.str.strip().str.replace(' ', '_')
df.dropna(subset=['CustomerID'], inplace=True)
df['CustomerID'] = df['CustomerID'].astype(int)
df.head(10)
```

```python
from sqlalchemy import create_engine
```

```python
# STEP 4: Connect to PostgreSQL
# Replace values below with your actual PostgreSQL config
db_user = 'postgres'
db_password = 'xxxxxx'
db_host = 'localhost'
db_port = '5432'
db_name = 'customers'

engine = create_engine(f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')
```

```python
df.to_sql('online_retail_raw', engine, index=False, if_exists='replace')
print("✅ Data loaded into PostgreSQL successfully")
```
